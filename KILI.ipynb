```python
import pandas as pd
from scipy.stats import spearmanr
```

```python
df1 = pd.read_csv('/content/NOUN 0.csv', sep=";")
df1
```

```python
rho, p = spearmanr(df1["Крюков"], df1["Шолохов"])

print(f'Nouns, Коэффициент Смирмана: {rho}, p: {p}')
```
