```python
import pandas as pd
from scipy.stats import mannwhitneyu
import matplotlib.pyplot as plt
```

```python
ipm = pd.read_csv("C:/Users/Ольга/Desktop/ipm.csv")
ipm.head()
```

```python
fig, axs = plt.subplots(nrows= 2 , ncols= 1, figsize=(15,10), sharey=True)
axs[0].hist(ipm['к'], histtype='step', fill=True)
axs[1].hist(ipm['ш'], histtype='step', fill=True)
```

H 0 : две популяции равны
H a : две популяции не равны

```python
stat, p = mannwhitneyu(ipm['к'], ipm['ш'])
print(f'Statistics={stat}, p={p}')
```

```python
alpha = 0.05
if p > alpha:
    print('fail to reject H0')
else:
    print('reject H0')
```
